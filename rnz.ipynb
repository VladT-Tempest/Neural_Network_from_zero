Taking advantage of the fact that within each layer of a neural network we have the same types of neurons -> the same activation function and we apply the same backpropagation formulas, we use vectorization and define each layer as a module, so we define a class in Python that is refer to a LAYER:

```python
# NETWORK LAYER CLASS

class neural_layer():

    def __init__(self, n_conn, n_neur, act_f):
        self.act_f = act_f # Activation function
        self.b = np.random.rand(1, n_neur) * 2 - 1 # bias, column vector initialized randomly and normalized (from -1 to 1)
        self.W = np.random.rand(n_conn, n_neur) * 2 - 1  # matrix
```

With this we can now create layers for our neural network, now we are going to create the activation functions:

Sigmoide --> $\sigma(x) = \frac{1}{1 + e^{-x}}$


```python
# ACTIVATION FUNCTIONS

sigm = (lambda x: 1 / (1 + np.e ** (-x)),
        lambda x: x * (1-x))  # This part is the derivative of the sigmoid function

_x = np.linspace(-5,5,100) 
print(_x)
#plt.xlim(-6,6)
#plt.ylim(-30, 4)
plt.plot(_x, sigm[0](_x))
#plt.plot(_x, sigm[1](_x))
```

```python
#  here we define the structure of our layers
l0 = neural_layer(p, 4, sigm)
l1 = neural_layer(4, 8, sigm)
# .......



def create_nn(topology, act_f):

    nn = [] # vector where we will contain all the layers of the neural network
    
    for l, layer in enumerate(topology[:-1]):
        nn.append(neural_layer(topology[l], topology[l+1], act_f))

    return nn
```

```python
topology = [p, 4, 8, 1] # structure/number of neurons in each layer

neural_net = create_nn(topology, sigm) # --> Neural network

l2_cost = (lambda Yp, Yr: np.mean((Yp - Yr) ** 2),   # We define our cost function
           lambda Yp, Yr: (Yp - Yr))                 # This second part is the derivative

def train(neural_net, X, Y, l2_cost, lr=0.5, train=True):

    out = [(None, X)]

    # forward pass
    """
   We take our input vector and we are going to pass it layer by layer, combining the operations that are performed in each of the neurons: weighted sum -> we multiply the input value x by w and add the bias and then we pass that weighted sum to a function activation
    """
    for l, layer in enumerate(neural_net):

        z = out[-1][1] @ neural_net[l].W + neural_net[l].b # -> weighted sum of first layer
        a = neural_net[l].act_f[0](z) # a -> layer 1 output

        out.append((z, a))
    
    print(l2_cost[0](out[-1][1], Y))

    if train:  # If the training is not activated, there is no training and we only obtain a prediction

        # Backward pass
        """ 
        Backpropagation is implemented with the partial derivatives that we need for the gradient descent algorithm, which are the parameters W and b with respect to the cost.
        """
        deltas = []

        for l in reversed(range(0, len(neural_net))):

            z = out[l+1][0]  # l+1 to work with hidden layers
            a = out[l+1][1]

            print(a.shape)

            if l == len(neural_net) - 1: # if we are in the last layer
                # Calculate last layer delta -> 1.
                deltas.insert(0, l2_cost[1](a, Y) * neural_net[l].act_f[1](a)) # Formula to calculate delta0
            else:
                # Calculate delta with respect to previous layer -> 2.
                 deltas.insert(0, deltas[0] @ _W.T * neural_net[l].act_f[1](a))


            _W = neural_net[l].W
                

            # Gradient descent            Combined with Backpropagation they would be training
            neural_net[l].b = neural_net[l].b - np.mean(deltas[0],axis=0, keepdims=True) * lr
            neural_net[l].W = neural_net[l].W - out[l][1].T @ deltas[0] * lr    # -> 3.

        return out[-1][1]


train(neural_net, X, Y, l2_cost, 0.5)
```

1. Computo de error de la última capa:

$\delta^L = \frac{\delta C}{\delta a^L} . \frac{\delta a^L}{\delta z^L}$

2. Retropropagamos el error a la capa anterior

$ \delta^{l-1} = W^l \delta^l . \frac{\delta a^{l-1}}{\delta z^{l-1}} $

3. Calculamos las derivadas de la capa usando el error:

$ \frac{\delta C}{\delta b^{l-1}} = \delta^{l-1}  \qquad  \frac{\delta C}{\delta w^{l-1}} = \delta^{l-1} a^{l-2} $

```python
# Testing the Neural Network !!
import time
from IPython.display import clear_output

neural_n = create_nn(topology, sigm)

loss = []

for i in range(2500):
    # Let's train the network!
    pY = train(neural_n, X, Y, l2_cost, lr=0.05)
    
    if i %  25 == 0:

        loss.append(l2_cost[0](pY, Y))
        
        res = 50

        _x0 = np.linspace(-1.5, 1.5, res)
        _x1 = np.linspace(-1.5, 1.5, res)

        _Y = np.zeros((res, res))

        for i0, x0 in enumerate(_x0):
            for i1, x1 in enumerate(_x1):
                _Y[i0, i1] = train(neural_n, np.array([[x0, x1]]), Y, l2_cost, train=False)

        plt.pcolormesh(_x0, _x1, _Y, cmap="coolwarm")
        plt.axis("equal")

        plt.scatter(X[Y[:,0] == 0, 0], X[Y[:,0] == 0, 1], c="skyblue")
        plt.scatter(X[Y[:,0] == 1, 0], X[Y[:,0] == 1, 1], c="salmon")

        clear_output(wait=True)
        plt.show()
        plt.plot(range(len(loss)), loss)
        plt.show()
        time.sleep(0.5)



```
